```python
class ProbDist(dict):
    """A Probability Distribution; an {outcome: probability} mapping."""
    def __init__(self, mapping=(), **kwargs):
        self.update(mapping, **kwargs)
        # Make probabilities sum to 1.0; assert no negative probabilities
        total = sum(self.values())
        for outcome in self:
            self[outcome] = self[outcome] / total
            assert self[outcome] >= 0
```

```python
#Question 1
def calculate_race_probabilities(driver_probs, team_drivers):
    
    # 1. Probability distribution for the Australia Grand Prix
    australia_gp_probs = driver_probs.copy()

    # 2. Probability distribution for winning both Australia and China Grand Prix (assuming independent events)
    dual_race_win_probs = {driver: prob**2 for driver, prob in driver_probs.items()}

    # 3. Probability for Red Bull Racing to win both races
    red_bull_team = team_drivers.get('Red Bull Racing', [])
    red_bull_dual_win_prob = sum(dual_race_win_probs[driver] for driver in red_bull_team)

    # 4. Probability for Red Bull Racing to win at least one of the two races
    red_bull_australia_prob = sum(australia_gp_probs[driver] for driver in red_bull_team)
    red_bull_at_least_one_win = 1 - (1 - red_bull_australia_prob) ** 2 

    # 5. Probability of Max Verstappen winning again in 2025
    mv_prob = australia_gp_probs.get('MV', 0)  # Default to 0 if 'MV' is not found
    total_other_drivers_prob = sum(prob for driver, prob in australia_gp_probs.items() if driver != 'MV')
    mv_win_2025 = mv_prob / total_other_drivers_prob if total_other_drivers_prob > 0 else 0  # Avoid division by zero

    return {
        'Australia GP Probabilities': australia_gp_probs,
        'Both Races Probabilities': dual_race_win_probs,
        'Red Bull Both Races Probability': red_bull_dual_win_prob,
        'Red Bull At Least One Race Probability': red_bull_at_least_one_win,
        'Max Verstappen Winning 2025 Probability': mv_win_2025
    }


q1_results = calculate_race_probabilities(AGP, team_mapping)

for key, value in q1_results.items():
    print(f"\n{key}:")
    if isinstance(value, dict):
        for driver, prob in value.items():
            print(f"{driver}: {prob:.4f}")
    else:
        print(f"{value:.4f}")
```

```python
#Question 2
def calculate_conditional_probabilities(driver_probs, team_drivers):
  
    team_probabilities = {}

    for team, drivers in team_drivers.items():
        # Probability of the team winning a single race (Australia GP)
        team_single_race_prob = sum(driver_probs[driver] for driver in drivers)
        
        # Probability of the team winning both Australia and China GP (assuming independence)
        team_dual_win_prob = team_single_race_prob ** 2
        
        # Probability of the team winning at least one of the two races
        team_at_least_one_win_prob = 1 - (1 - team_single_race_prob) ** 2  # Complement rule

        # Probability of winning both races given the team wins at least one
        conditional_dual_win_prob = (team_dual_win_prob / team_at_least_one_win_prob) if team_at_least_one_win_prob > 0 else 0

        # Store results for the team
        team_probabilities[team] = {
            'Winning China GP given win in Australia GP': team_single_race_prob,
            'Winning both races given at least one win': conditional_dual_win_prob
        }

    return team_probabilities


q2_results = calculate_conditional_probabilities(AGP, team_mapping)


for team, probabilities in q2_results.items():
    print(f"\nTeam: {team}")
    for description, prob in probabilities.items():
        print(f"  {description}: {prob:.4f}")
```

```python
#Question 3
def ferrari_race_win_probabilities(driver_probs, team_drivers):

    # Ferrari drivers
    ferrari_team = team_drivers.get('Ferrari', [])

    # Probability of a sunny race day
    sunny_day_chance = 1 / 5  

    # Probability of Ferrari winning a single race (Australia GP)
    ferrari_win_single_race = sum(driver_probs[driver] for driver in ferrari_team)

    # Probability of Ferrari winning both Australia and China GP when both race days are sunny
    ferrari_dual_win_sunny = (ferrari_win_single_race * sunny_day_chance) ** 2

    return {
        'Ferrari Win Probability (Single Race)': ferrari_win_single_race,
        'Ferrari Winning Both Races on Sunny Days': ferrari_dual_win_sunny
    }



q3_results = ferrari_race_win_probabilities(AGP, team_mapping)

for description, probability in q3_results.items():
    print(f"{description}: {probability:.4f}")
```

```python
#Question 4
def calculate_fashion_show_probabilities(total_events=10, total_participants=20):
    
    # Probability of meeting Max Verstappen at a randomly chosen event
    prob_meeting_mv = 1 / total_events

    # Probability of meeting any driver if all drivers attend the same show
    prob_meeting_any_driver_same_show = min(1, total_participants / total_events)

    # Probability of a single driver attending a specific event
    prob_driver_attends_event = 1 / total_events

    # Probability of meeting at least one driver across all events
    prob_meeting_any_driver_any_event = 1 - (1 - prob_driver_attends_event) ** total_participants

    return {
        'Meeting Max Verstappen': prob_meeting_mv,
        'Meeting Any Driver (All in One Event)': prob_meeting_any_driver_same_show,
        'Meeting Any Driver (Across Events)': prob_meeting_any_driver_any_event
    }



q4_results = calculate_fashion_show_probabilities()

for scenario, probability in q4_results.items():
    print(f"{scenario}: {probability:.4f}")
```
